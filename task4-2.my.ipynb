```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize']=(12,5)
```

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
```

```python
digit=load_digits()
digit
```

```python
img=digit.images
```

```python
img.shape
```

```python
plt.imshow(img[0])
```

```python
X=img.reshape(-1,64)
X.shape
```

```python
tsne=TSNE(n_components=2,perplexity=30.)
```

```python
tsne.fit(X)
```

```python
Z=tsne.embedding_
```

```python
Z.shape
```

```python
y=digit.target
```

```python
plt.scatter(Z[:,0],Z[:,1],c=y)
```

```python
perplex=[1,3,5,10,30,50]
```

```python
for p in perplex:
    tsne=TSNE(n_components=2,perplexity=p)
    tsne.fit(X)
    Z=tsne.embedding_
    fit,ax=plt.subplots(1,1)
    ax.scatter(Z[:,0],Z[:,1],c=y)
    ax.set_title('Perplexity{}'.format(p))
```
