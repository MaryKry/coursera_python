```python
import time
import datetime
```

```python
citibike['start_hour'] = citibike['starttime'].apply(lambda x: datetime.datetime.fromtimestamp(
                          time.mktime(datetime.datetime.strptime(x.strip(), "%Y-%m-%d %H:%M:%S.%f").timetuple())).hour)
                                                                 
citibike['end_hour'] = citibike['stoptime'].apply(lambda x: datetime.datetime.fromtimestamp(
    time.mktime(datetime.datetime.strptime(x.strip(), "%Y-%m-%d %H:%M:%S.%f").timetuple())).hour) 
                                           
citibike[(citibike.start_hour.isin([18, 19, 20]))]['start station id'].value_counts().head()                                           
```
